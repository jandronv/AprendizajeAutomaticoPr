```python
print('Crea un array de 0 de tamaño 10')
print(np.zeros(10, dtype=int),'\n')

print('Crea un array de 3x5 de 1s')
print(np.ones((3, 5), dtype=float),'\n')

print('Crea un array de 3x5 de 1s')
print(np.ones((3, 5), dtype=float),'\n')

print('Crea un array de 3x5 de 3.14')
print(np.full((3, 5), 3.14),'\n')

print('Crea un array secuencial de 0 a 20')
print(np.arange(0,20,2),'\n')

print('Crea un array de 5 con valores entre 0 y 1')
print(np.linspace(0,1,5),'\n')

print('Crea un array multidimensional random/normal/randint')
print(np.random.random((3, 3)),'\n')

print('Manipular la dimensionalidad con reshape')
grid = np.arange(1, 10).reshape((3, 3))
print(grid)
```

### Array Slicing x[start:stop:step]

```python
x = np.arange(10)

#Primeros elementos
print(x[:5],'\n')

#Elementos desde el 5
print(x[5:],'\n')

#Elementos intermedios
print(x[2:5],'\n')

#Elementos de 2 en 2
print(x[::2],'\n')

print(x2[:2, :3],'\n')

print(x2[:, 0],'\n')

#Crear copias
x2_sub_copy = x2[:2,:2].copy()
print(x2_sub_copy)
```

```python
import pandas as pd
import numpy as np

data = pd.read_csv('data/president_heights.csv')
reviews = pd.read_csv('data/winemag-data-130k-v2.csv')
heights = np.array(data['height(cm)'])
print(heights)
```

```python
print("Mean height: ", heights.mean())
print("Standard deviation:", heights.std())
print("Minimum height: ", heights.min())
print("Maximum height: ", heights.max())

print("25th percentile: ", np.percentile(heights, 25))
print("Median: ", np.median(heights))
print("75th percentile: ", np.percentile(heights, 75))
```

```python
import matplotlib.pyplot as plt
import seaborn; seaborn.set()

plt.hist(heights)
plt.title('Height Distribution of US Presidents')
plt.xlabel('height (cm)')
plt.ylabel('number');
```
