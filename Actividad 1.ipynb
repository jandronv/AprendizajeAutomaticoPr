```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor(n_estimators=300,max_depth=90, max_features=10, min_samples_leaf=3, min_samples_split=8)
# Train the model on training data
rfr.fit(train_features, train_labels)
```

```python
# Use the forest's predict method on the test data
predictions = rfr.predict(test_features)

# Calculate the absolute errors
errors = abs(predictions - test_labels)
```

```python
from sklearn.metrics import mean_squared_log_error
from sklearn.metrics import mean_absolute_error, mean_squared_error


# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / test_labels)

# Calculate and display accuracy
accuracy = 100 - np.mean(mape)

print('Accuracy:', round(accuracy, 2), '%.')

print('MAE:',mean_absolute_error(predictions,test_labels))
print('MSE:',mean_squared_error(predictions,test_labels))
print('RMSE:',np.sqrt(mean_squared_error(predictions,test_labels)))
```

# Problema de clasificación

```python
from sklearn.tree import DecisionTreeClassifier
#Creamos una nueva columna con las calses de precio
house_data_frame.loc[(house_data_frame['SalePrice']<=100000.0), 'SalePriceClass'] = 0
house_data_frame.loc[(house_data_frame['SalePrice']>100000.0) & (house_data_frame['SalePrice']<500000.0), 'SalePriceClass'] = 1
house_data_frame.loc[(house_data_frame['SalePrice']>=500000.0), 'SalePriceClass'] = 2
```

```python
from sklearn.model_selection import train_test_split
y = house_data_frame['SalePriceClass']

X = house_data_frame.drop(['SalePriceClass', 'SalePrice'],axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
# Creamos el arbol de decisión
dt_clf = DecisionTreeClassifier(max_depth=9,min_samples_leaf=2)

# Entrenamos
dt_clf = dt_clf.fit(X_train,y_train)

#predecimos
y_pred = dt_clf.predict(X_test)
```

```python
from sklearn import metrics

# Mostrar confusion matrix
print(metrics.confusion_matrix(y_test, y_pred))

# Mostrar precision and recall, junto con el resto de métricas
print(metrics.classification_report(y_test, y_pred, digits=3))
```

```python
from sklearn.ensemble import RandomForestClassifier


rf_clf = RandomForestClassifier(n_estimators=300,max_depth=90, max_features=10, min_samples_leaf=3, min_samples_split=8)

rf_clf.fit(X_train,y_train)

y_pred = rf_clf.predict(X_test)
```

```python
# Mostrar confusion matrix
print(metrics.confusion_matrix(y_test, y_pred))

# Mostrar precision and recall, junto con el resto de métricas
print(metrics.classification_report(y_test, y_pred, digits=3))
```
